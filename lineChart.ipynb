##### Line Plots

```Line plot for immigrants from Haiti```

````python
#extracting data for Haiti
haiti = df_can.loc['Haiti', years]
haiti.head()
````

````python
#using plot function to plot a line chart
haiti.plot(kind='line')

#pandas automatically populate x axis with index values and 
# y-axis with column values
````

````python
#to give title and labels to line chart
haiti.index = haiti.index.map(int)  #converting index value to int
haiti.plot(kind='line')
plt.title('Immigrants from Haiti')
plt.xlabel('Years')
plt.ylabel('Numbers of immigrants')
plt.text(2000, 6000, '2010 Earthquake')
plt.show() #to show update to figures
````


*Since the x-axis (years) is type 'integer', we specified x as a year. The y axis (number of immigrants) is type 'integer', so we can just specify the value y = 6000.*
```python
    plt.text(2000, 6000, '2010 Earthquake') # years stored as type int
```
*If the years were stored as type 'string', we would need to specify x as the index position of the year. Eg 20th index is year 2000 since it is the 20th year with a base year of 1980.*
```python
    plt.text(20, 6000, '2010 Earthquake') # years stored as type int
```

```Line chart for comparison of immigrants from India and China```

````python
df_IC = df_can.loc[['India', 'China'], years]
df_IC
````

````python
#We can't simply use plot function to plot chart, 
#as index of df_IC is 'India' and 'China' 
#and we need years to be index
df_IC = df_IC.transpose()
df_IC
````

````python
#plotting line chart for comparison India and China immigrants
df_IC.index = df_IC.index.map(int) 
df_IC.plot(kind='line')
plt.title('Immigrants from China and India')
plt.xlabel('Years')
plt.ylabel('Numbers of immigrants')
plt.show()
````
